# PyTorch Library

The power of deep learning is enabled by libraries such as PyTorch and TensorFlow that simplify 
- Synthetic data generation
- Defining the machine learning model, loss function and optimizer
- Model training loop with gradient descent
- Visualization of loss and results


```python
# Linear Regression with PyTorch
# Description: Introductory example for training a linear regression model using PyTorch

import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Set seed for reproducibility
torch.manual_seed(42)
```

## 1. Generate synthetic linear data

```python
# Generate PyTorch array with data of y = 2*x + 3 + noise
X = torch.unsqueeze(torch.linspace(0, 10, 100), dim=1)  # shape: (100, 1)
y = 2 * X + 3 + torch.randn(X.size()) * 2

plt.scatter(X.numpy(), y.numpy())
plt.title("Generated data")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

## 2. Define a model

```python
# Define a simple linear layer (y = wx + b)
model = nn.Linear(in_features=1, out_features=1)
```

## 3. Define loss function and optimizer

```python
# Use MSE loss function and stochastic gradient decent optimizer
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

## 4. Training loop

```python
# Loop over epochs and change parameters using backpropagation of the loss function and gradient descent
epochs = 200
loss_history = []

# Training loop
for epoch in range(epochs):
    model.train()

    # Forward pass
    y_pred = model(X)

    # Compute loss
    loss = loss_fn(y_pred, y)
    loss_history.append(loss.item())

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Print loss every 20 epochs
    if epoch % 20 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```

## 5. Visualization of loss and results

```python
# Plot the loss curve
plt.plot(loss_history)
plt.title("Training Loss Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.show()
```

```python

# Plot fitted line
predicted = model(X).detach().numpy()
plt.scatter(X.numpy(), y.numpy(), label='Data')
plt.plot(X.numpy(), predicted, color='red', label='Fitted line')
plt.title("Linear Fit with PyTorch")
plt.legend()
plt.show()
```

```python

# View learned parameters
print("Learned weight:", model.weight.item())
print("Learned bias:", model.bias.item())
```
